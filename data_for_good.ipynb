```python
import csv
from matplotlib.pyplot import *

with open('/home/jupyter-user-1/team-28/FFG_Hackathon_Country_Level_Data.csv') as country_file:
    cvs_reader = csv.DictReader(country_file)
    
    country_year, country_name, country_nuclear, country_fossil_fuel, country_emissions = [], [], [], [], []
    china_year, china_nuclear, china_fossil_fuel = [], [], []
    iceland_year, iceland_nuclear, iceland_fossil_fuel = [], [], []
    higher_fossil_fuel_countries_set, lower_fossil_fuel_countries_set = set(), set()
    
    for row in cvs_reader:
        
        country_year.append(row['Year'])
        country_name.append(row['Country Name'])
        country_nuclear.append(row['EG.USE.COMM.CL.ZS'])
        country_fossil_fuel.append(row['EG.USE.COMM.FO.ZS'])
        
for index in range(1310):
        
    # Check for any empty or NULL values 
    if country_nuclear[index] != 0 and country_nuclear[index] != '' and country_fossil_fuel[index] != 0 and country_fossil_fuel[index] != '':
            
        # Countries with lower nuclear energy and higher fossil fuel energy
        if country_nuclear[index] < country_fossil_fuel[index]:
            
            # If the country is not already recorded
            if country_name[index] not in higher_fossil_fuel_countries_set:
            
                higher_fossil_fuel_countries_set.add(country_name[index])
                
        # Countries with higher nuclear energy and lower fossil fuel energy
        if country_nuclear[index] > country_fossil_fuel[index]:
            
            # If the country is not already recorded
            if country_name[index] not in lower_fossil_fuel_countries_set:
            
                lower_fossil_fuel_countries_set.add(country_name[index])
                
print(len(higher_fossil_fuel_countries_set), "Countries with higher fossil fuel energy than nuclear energy (%), including", higher_fossil_fuel_countries_set, "\n\n")
print(len(lower_fossil_fuel_countries_set), "Countries with higher nuclear energy than fossil fuel energy (%), including", lower_fossil_fuel_countries_set, "\n\n")
    
for index in range(1310):
        
    if country_name[index] == "China":
            
        china_year.append(country_year[index])
        china_nuclear.append(country_nuclear[index])
        china_fossil_fuel.append(country_fossil_fuel[index])
            
    if country_name[index] == "Iceland":
            
        iceland_year.append(country_year[index])
        iceland_nuclear.append(country_nuclear[index])
        iceland_fossil_fuel.append(country_fossil_fuel[index])
            
china_dict = {"Year": china_year, "Nuclear Energy": china_nuclear, "CO2 Emissions": china_fossil_fuel}
iceland_dict = {"Year": iceland_year, "Nuclear Energy": iceland_nuclear, "CO2 Emissions": iceland_fossil_fuel}

fig, axs1 = subplots(2, 1)

axs1[0].scatter(china_year, china_nuclear)
axs1[0].set_title("China: Alternative and Nuclear Energy")
axs1[0].set(xlabel = "Year", ylabel = "Alternative and Nuclear Energy (%)")
axs1[1].scatter(china_year, china_fossil_fuel)
axs1[1].set_title("China: Fossil Fuel Energy")
axs1[1].set(xlabel = "Year", ylabel = "Fossil Fuel Energy Consumption (%)")

subplots_adjust(bottom = 0, top = 1, wspace = 0.2, hspace = 0.9)

fig, axs2 = subplots(2, 1)

axs2[0].scatter(iceland_year, iceland_nuclear)
axs2[0].set_title("Iceland: Alternative and Nuclear Energy")
axs2[0].set(xlabel = "Year", ylabel = "Alternative and Nuclear Energy (%)")
axs2[1].scatter(iceland_year, iceland_fossil_fuel)
axs2[1].set_title("Iceland: Fossil Fuel Energy")
axs2[1].set(xlabel = "Year", ylabel = "Fossil Fuel Energy Consumption (%)")

subplots_adjust(bottom = 0, top = 1, wspace = 0.2, hspace = 0.9)

show()
```
